```
# Step 2: Set up the data folder with the sample file (50 candidates)
import os, json
os.makedirs('data', exist_ok=True)

# The sample file is already in the repo
# Convert sample_candidates.json to jsonl format for the pipeline
with open('data/sample_candidates.json') as f:
    candidates = json.load(f)

with open('data/candidates.jsonl', 'w') as f:
    for c in candidates:
        f.write(json.dumps(c) + '\n')

print(f'Loaded {len(candidates)} sample candidates')
```
